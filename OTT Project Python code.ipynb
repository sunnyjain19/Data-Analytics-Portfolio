```python
df.dropna(subset=['duration', 'rating', 'date_added'], inplace=True)
```

```python
df.shape
```

```python
df.drop(columns =['show_id','description'],inplace=True)
```

```python
df.head()
```

```python
df.type.value_counts()
```

```python
df['director']=df['director'].fillna('Not Mentioned')
```

```python
df['cast']=df['cast'].fillna('Unknown')
```

```python
df['country']=df['country'].fillna('Not Mentioned')
```

```python
df.isna().sum()
```

```python
df['country'].unique()
```

```python
df['country'] = df['country'].str.split(',').str[0]
```

```python
df['country'].unique()
```

```python
df['country'].value_counts()
```

```python
df['date_added']=df['date_added'].str.replace(',','')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],format = 'mixed')
```

```python
df.info()
```

```python
df['year_added']=df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month_name()
df['day_added']=df['date_added'].dt.day_name()
```

```python
df.head()
```

```python
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']

df['month_added'] =pd.Categorical(df['month_added'],categories = month_order, ordered =True)
df['day_added'] =pd.Categorical(df['day_added'],categories = day_order, ordered =True)
```

```python
df['rating'].unique()
```

```python
df['duration'].unique()
```

```python
df['lead_cast'] = df['cast'].str.split(",").str[0]
df['supporting_cast'] = df['cast'].str.split(',').str[1]
```

```python
df['supporting_cast'].isna().sum()
```

```python
df['supporting_cast']=df['supporting_cast'].fillna("N/A")
```

```python
df.head()
```

```python
df['genre']=df['listed_in'].str.split(",").str[0] 
```

```python
df.drop(columns=['listed_in','cast'],inplace = True)
```
